```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_excel("datos_mex.xlsx", skiprows=3)

df.columns = ['Entidad','1990','1995','2000','2005','2010','2020']
df.head()
```

```python
df_melt = df.melt(id_vars='Entidad', var_name='Año', value_name='Acceso')

fig = px.line(
    df_melt,
    x='Año',
    y='Acceso',
    color='Entidad',
    title='Evolución del acceso a electricidad por entidad en México'
)

fig.show()
```

```python
top5_1990 = df.sort_values('1990', ascending=False).head(5)

fig = px.bar(
    top5_1990,
    x='Entidad',
    y='1990',
    title='Top 5 estados con mayor acceso a electricidad en 1990'
)

fig.show()
```

```python
top5_2020 = df.sort_values('2020', ascending=False).head(5)

df_top = df[df['Entidad'].isin(top5_2020['Entidad'])]
df_top_melt = df_top.melt(id_vars='Entidad', var_name='Año', value_name='Acceso')

fig = px.line(
    df_top_melt,
    x='Año',
    y='Acceso',
    color='Entidad',
    title='Evolución del Top 5 estados (según 2020)'
)

fig.show()
```
